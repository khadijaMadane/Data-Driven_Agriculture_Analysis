```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.colorbar import ColorbarBase
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Charger le GeoDataFrame
geojson_path = '/content/maroc (1).geojson'
gdf = gpd.read_file(geojson_path)

# Charger le DataFrame des données
dataframe = pd.read_csv("/content/cheptel-reg.csv", sep=",", encoding="ISO-8859-1")
print(dataframe)
# Fusionner les GeoDataFrame et DataFrame
merged = gdf.merge(dataframe, left_on='region', right_on='Zone-geographique', how='left')

# Remplacer les valeurs NaN par une valeur par défaut (par exemple, 0)
merged['2017'].fillna(0, inplace=True)

# Ajuster ces valeurs selon votre plage de données
cmap = plt.get_cmap('YlOrRd')
merged['2017'] = pd.to_numeric(merged['2017'], errors='coerce')
norm = Normalize(vmin=0, vmax=merged['2017'].max())
levels = [0, 1000000, 2000000, 3000000, merged['2017'].max()]

# Créer une figure et un axe
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.set_title("L'Effectif du Cheptel par région au Maroc pour l'année 2017")

# Afficher le fond de la carte du Maroc
gdf.plot(ax=ax, color='lightgray')

# Ajouter les noms des régions
for x, y, label in zip(merged.geometry.centroid.x, merged.geometry.centroid.y, merged['Zone-geographique']):
    ax.text(x, y, label, fontsize=8, ha='center', va='center')

# Afficher les données avec une couleur différente pour chaque intervalle
merged.plot(column='2017', cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.8', legend=False, missing_kwds={'color': 'gray'})

# Ajouter une légende explicative
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cb = ColorbarBase(cax, cmap=cmap, norm=norm, ticks=levels, orientation='vertical')

# Étiquettes explicatives pour chaque seuil
cb.set_label("L'Effectif du Cheptel", labelpad=10)

plt.show()
```
